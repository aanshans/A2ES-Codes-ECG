```python
# Load packages
import neurokit2 as nk                #NEUROKIT2
import pandas as pd                   #PANDA
import matlab.engine                  #MATLAB ENGINE (Matlab 2020b with installed Matlab Python integration)

#............Input RAW ECG Signal...................
raw = pd.read_csv('C:/Users/ANAS/Desktop/Anas Here/DATASET DREAMER/FIL.csv')    #read from .csv raw ECG signal, with 1 column & header 'ECG'

#Extract HRV using Neurokit2
peaks, info = nk.ecg_peaks(raw["ECG"], sampling_rate=256)                                   #detect R peaks
hrv_indices = nk.hrv(peaks, sampling_rate=256, show=False)                                  #extract HRV
hrv_indices.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DR_NK.csv', mode='a', header=False)   #save to .csv file 

#Extract HR using AuBT
eng = matlab.engine.start_matlab()          #start Matlab engine

xhead_raw = raw[1:]                     #remove 'ECG' header        
py_var = xhead_raw.values.tolist()      #convert datatype from panda dataframe to python list
mat_var = matlab.single(py_var)         #convert datatype from python list to Matlab matrix
extracted, labels = eng.aubt_extractFeatECG(mat_var, nargout=2)    #extract HR using AuBT in Matlab. Sampling rate/freq/hz need to change in the Matlab code if necessary
df_extracted = pd.DataFrame(extracted)      #convert datatype from Matlab matrix to panda dataframe
df_extracted.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DR_AU.csv', mode='a', header=False)  #save to .csv file

eng.quit()  #stop Matlab engine



#................FILTER NEUROKIT.....................
signals = nk.ecg_clean(xhead_raw, sampling_rate=256, method="neurokit")

f1 = pd.DataFrame(signals, columns=['ECG']) 

#NKfilterNK
peaks, info = nk.ecg_peaks(f1["ECG"], sampling_rate=256)
hrv_indices = nk.hrv(peaks, sampling_rate=256, show=False)
hrv_indices.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF1_NK.csv', mode='a', header=False)

#AUfilterNK
eng = matlab.engine.start_matlab()

xhead_raw1 = f1[1:]
py_var1 = xhead_raw1.values.tolist()
mat_var1 = matlab.single(py_var1)
extracted1, labels = eng.aubt_extractFeatECG(mat_var1, nargout=2)
df_extracted1 = pd.DataFrame(extracted1)
df_extracted1.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF1_AU.csv', mode='a', header=False)

eng.quit()


#................FILTER BIOSPPY.....................
signal1 = nk.ecg_clean(xhead_raw, sampling_rate=256, method="biosppy")

f2 = pd.DataFrame(signal1, columns=['ECG']) 

#NKfilterBIO
peaks, info = nk.ecg_peaks(f2["ECG"], sampling_rate=256)
hrv_indices = nk.hrv(peaks, sampling_rate=256, show=False)
hrv_indices.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF2_NK.csv', mode='a', header=False)

#AUfilterBIO
eng = matlab.engine.start_matlab()

xhead_raw2 = f2[1:]
py_var2 = xhead_raw2.values.tolist()
mat_var2 = matlab.single(py_var2)
extracted2, labels = eng.aubt_extractFeatECG(mat_var2, nargout=2)
df_extracted2 = pd.DataFrame(extracted2)
df_extracted2.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF2_AU.csv', mode='a', header=False)

eng.quit()


#................FILTER PAN.....................
signal2 = nk.ecg_clean(xhead_raw, sampling_rate=256, method="pantompkins1985")

f3 = pd.DataFrame(signal2, columns=['ECG']) 

#NKfilterPAN
peaks, info = nk.ecg_peaks(f3["ECG"], sampling_rate=256)
hrv_indices = nk.hrv(peaks, sampling_rate=256, show=False)
hrv_indices.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF3_NK.csv', mode='a', header=False)

#AUfilterPAN
eng = matlab.engine.start_matlab()

xhead_raw3 = f3[1:]
py_var3 = xhead_raw3.values.tolist()
mat_var3 = matlab.single(py_var3)
extracted3, labels = eng.aubt_extractFeatECG(mat_var3, nargout=2)
df_extracted3 = pd.DataFrame(extracted3)
df_extracted3.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF3_AU.csv', mode='a', header=False)

eng.quit()


#................FILTER HAM.....................
signal3 = nk.ecg_clean(xhead_raw, sampling_rate=256, method="hamilton2002")

f4 = pd.DataFrame(signal3, columns=['ECG']) 

#NKfilterHAM
peaks, info = nk.ecg_peaks(f4["ECG"], sampling_rate=256)
hrv_indices = nk.hrv(peaks, sampling_rate=256, show=False)
hrv_indices.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF4_NK.csv', mode='a', header=False)

#AUfilterHAM
eng = matlab.engine.start_matlab()

xhead_raw4 = f4[1:]
py_var4 = xhead_raw4.values.tolist()
mat_var4 = matlab.single(py_var4)
extracted4, labels = eng.aubt_extractFeatECG(mat_var4, nargout=2)
df_extracted4 = pd.DataFrame(extracted4)
df_extracted4.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF4_AU.csv', mode='a', header=False)

eng.quit()


#................FILTER ELG.....................
signal4 = nk.ecg_clean(xhead_raw, sampling_rate=256, method="elgendi2010")

f5 = pd.DataFrame(signal4, columns=['ECG']) 

#NKfilterELG
peaks, info = nk.ecg_peaks(f5["ECG"], sampling_rate=256)
hrv_indices = nk.hrv(peaks, sampling_rate=256, show=False)
hrv_indices.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF5_NK.csv', mode='a', header=False)

#AUfilterELG
eng = matlab.engine.start_matlab()

xhead_raw5 = f5[1:]
py_var5 = xhead_raw5.values.tolist()
mat_var5 = matlab.single(py_var5)
extracted5, labels = eng.aubt_extractFeatECG(mat_var5, nargout=2)
df_extracted5 = pd.DataFrame(extracted5)
df_extracted5.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF5_AU.csv', mode='a', header=False)

eng.quit()


#................FILTER ENG.....................
signal5 = nk.ecg_clean(xhead_raw, sampling_rate=256, method="engzeemod2012")

f6 = pd.DataFrame(signal5, columns=['ECG']) 

#NKfilterENG
peaks, info = nk.ecg_peaks(f6["ECG"], sampling_rate=256)
hrv_indices = nk.hrv(peaks, sampling_rate=256, show=False)
hrv_indices.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF6_NK.csv', mode='a', header=False)

#AUfilterENG
eng = matlab.engine.start_matlab()

xhead_raw6 = f6[1:]
py_var6 = xhead_raw6.values.tolist()
mat_var6 = matlab.single(py_var6)
extracted6, labels = eng.aubt_extractFeatECG(mat_var6, nargout=2)
df_extracted6 = pd.DataFrame(extracted6)
df_extracted6.to_csv('C:/Users/ANAS/Desktop/Anas Here/D/DF6_AU.csv', mode='a', header=False)

eng.quit()
```
